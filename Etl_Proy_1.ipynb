```python
import pandas as pd
import json
import ast
import matplotlib.pyplot  as plt
import warnings
warnings.filterwarnings("ignore")
```

```python
# Se especifica la ruta a los archivos y se carga el dataset
from pathlib import Path

data = []

file = Path('australian_user_reviews.json')  
with file.open('r', encoding='utf-8') as f:  
    for line in f: # read
        data.append(ast.literal_eval(line))
user_review = pd.DataFrame(data)
f.close()
```

```python
#Se eliminan valores Nulos para poder interpretar el dataset
user_review = user_review.dropna(how='all') 
user_review.to_csv(r'user_review.csv',index=False)
```

```python

# Se desanida el dataset
# Se crea el diccionario del nuevo df expandido y lo llevamos a csv
review_exp = []

for i,r in user_review.iterrows():
    user_id  = r['user_id']
    user_url = r['user_url']

    review_list = r['reviews']

    for x in review_list:
        
        nuevo_item = {    'user_id'  : user_id ,
                        'user_url' : user_url,
                        **x
                    }
        review_exp.append(nuevo_item)

df_review_exp = pd.DataFrame(review_exp)
df_review_exp.to_csv(r'user_review.csv',index=False)
```

```python
# Se especifica la ruta a los archivos y se carga el dataset
data = []
file = Path('australian_users_items.json')  
with file.open('r', encoding='utf-8') as f: 
    for line in f: 
        data.append(ast.literal_eval(line))
user_items = pd.DataFrame(data)
f.close()
```

```python
#Se eliminan valores Nulos para poder interpretar el dataset
user_items = user_items.dropna(how='all') 
user_items.to_csv(r'user_items.csv',index=False)
```

```python
# Se desanida el dataset
# Se crea el diccionario del nuevo df expandido y lo llevamos a csv

items_exp = []

for i,r in user_items.iterrows():
    user_id  = r['user_id']
    steam_id = r['steam_id']
    user_url = r['user_url']
    items_count = r['items_count']
    items_list = r['items']

    for x in items_list:
        
        nuevo_item = {    'user_id'  : user_id,
                        'items_count':items_count,
                        'steam_id' : steam_id,
                        'user_url' : user_url,
                        **x  
                    }
        items_exp.append(nuevo_item)

df_items_exp = pd.DataFrame(items_exp)
df_items_exp.to_csv(r'user_items.csv',index=False)
```

```python
# Se especifica la ruta a los archivos y se carga el dataset
# Se crea el diccionario del nuevo df y lo llevamos a csv
data = []

file = Path('output_steam_games.json')  
with file.open('r', encoding='utf-8') as f: 
    for line in f: 
        data.append(json.loads(line))
user_games = pd.DataFrame(data)
f.close()
```

```python
#Se eliminan valores Nulos para poder interpretar el dataset
user_games = user_games.dropna(how='all') 
user_games.to_csv(r'user_games.csv',index=False)
```
